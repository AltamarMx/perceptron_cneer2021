```python
class maquinita:

  def problema(self,x,w,b):
    return np.dot(x,w) + b

  def sigmoide(self,x):
    return 1/(1 + np.exp(-x))
  def costo(self, y,Y):
    return (y-Y)**2

  def punto(self, x,w,b,Y):
    y = self.sigmoide(self.problema(x,w,b))
    C = self.costo(y,Y)
    return C
```
